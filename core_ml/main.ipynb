```python
import numpy as np
from scipy.stats import norm
N = 10000000
p = 0.25
X = np.random.uniform(low=0, high=1, size=N)
Z = np.random.uniform(low=0, high=1, size=N)
I = np.random.uniform(low=0, high=1, size=N) <= p
Y = I * X + (1 - I) * Z
print(np.corrcoef(X, Y))
F1 = norm.ppf(X)
F2 = norm.ppf(Y)
print(np.corrcoef(F1, F2))
```

```python
N = 40000000
X = np.random.uniform(low=0, high=1, size=N)
Y = (np.random.uniform(low=1, high=2, size=N) + X ) / 3
print(np.corrcoef(X, Y))
print(np.var(Y), 1/12)
F1 = norm.ppf(X)
F2 = norm.ppf(Y)
print(np.corrcoef(F1, F2))
```
